```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


path = '/home/pasha/SatoshiVault/BTCUSDT-1m-2024-01-03.csv'

df = pd.read_csv(path)
df.columns = list(range(df.shape[1]))
df.iloc[:, 0] = pd.to_datetime(df.iloc[:, 0], unit='ms')

df.set_index(0, inplace=True)
df = df.rename_axis('timestamp')
price = df.iloc[:, 0]
returns = price.pct_change()* 100
volatility = returns.std()

volatility
```

```python
# Plot the returns using Matplotlib
plt.plot(returns.index, returns.values, label='Daily Returns')
plt.xlabel('Timestamp')
plt.ylabel('Daily Returns')
plt.title('Daily Returns of Column "n"')
plt.legend()
plt.show()
```
